```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mat=np.random.randint(1,200,(100,30))
mat
```

```python
mat=mat.astype('float64')
for i in range (mat.shape[0]):
    for j in range( mat.shape[1]):
        if mat[i][j] >= 10 and mat[i][j]<=60:
            mat[i][j]=np.nan
```

```python
mat
```

```python
for j in range( mat.shape[1]):
    colsum=np.average(np.nan_to_num(mat[:,j]))
    for i in range( mat.shape[0]):
        if np.isnan(mat[i][j]):
            mat[i][j]=colsum
```

```python
mat
```

```python
df=pd.DataFrame(mat)
cor=df.corr()
```

```python
sns.set(rc={'figure.figsize':(22,20)})
sns.heatmap(cor)
```

```python
cor
```

```python
count=0
for i in range(cor.shape[0]):
    for j in range(i+1,cor.shape[1]):
        if cor[j][i]<=.7:
            count+=1
count
```

```python
normalmat=mat
for i in range (normalmat.shape[1]):
    colmax=np.max(normalmat[:,i])
    colmin=np.min(normalmat[:,i])
    normalmat[:,i]=(normalmat[:,i]-colmin)/(colmax-colmin)
```

```python
normalmat=normalmat*10
```

```python
normalmat
```

```python
normalmat[normalmat<=5.0]=0
normalmat[normalmat>5]=1
# < 5 has to be first
```
