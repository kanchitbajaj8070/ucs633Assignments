```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(5)+1
y=[i*i for i in x]
z=[pow(i,3) for i in x]
```

# Line plot 

```python
plt.plot(x,y,color='blue',label="x square graph")
plt.plot(x,z,color='r',label=" y square graph")
plt.title( " X^2 ans X^3 graph")
plt.xlabel("x----->")
plt.ylabel("y------>")
plt.legend()
plt.show()
```

```python
plt.show()
```

```python
plt.plot(x,y,color='blue',label="x square graph",marker='o',markerfacecolor='red',linestyle="dashed",linewidth=2,markersize=4)
```

# Scatter plots 

```python
plt.scatter(x,y,c='purple',marker='*')
plt.scatter(x,z,color='red',marker='^')
plt.show()
```

# to denote each element on x axis

```python
plt.figure(figsize=(10,10))
a=np.array([1,2,3])*2
a
heights=[4,5,6] 
plt.bar(a,heights,width=.25,label="price in current year",tick_label=["gold","silver","copper"])# current year
heights2=[5,7,8]
plt.bar(a +0.25, heights2,width=0.25,color='red',label=" next year prices")# next year|
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10,10))
a=np.array([1,2,3])*2
a
heights=[4,5,6] 
plt.bar(a,heights,width=.25,label="price in current year",color='yellow',tick_label=["gold","silver","copper"])# current year
heights2=[5,7,8]
plt.bar(a +0.25, heights2,width=0.25,color='red',label=" next year prices",)# next year|
plt.legend()
plt.show()
```

# setting limits

```python

a=np.array([1,2,3])*2
a
heights=[4,5,6] 
plt.bar(a,heights,width=.25,label="price in current year",color='yellow',tick_label=["gold","silver","copper"])# current year
heights2=[5,7,8]
plt.bar(a +0.25, heights2,width=0.25,color='red',label=" next year prices",)# next year|
plt.legend()
plt.xlim(2,10)
plt.ylim(0,8)
plt.show()
```

# Pie chart 

```python
plt.pie?
```

```python
weightage=[20,10,4.5,6]
subjects=["maths","phy","chem","english"]
```

```python
plt.pie(weightage,labels=subjects)
plt.show()
```

```python
plt.pie(weightage,labels=subjects,explode=(0,0,0.2,0))# to show one of the subjects seperately
plt.show()
```

```python
plt.pie(weightage,labels=subjects,autopct="%1.1f%%",shadow=True,explode=(0,0.1,0.3,0)) # showing % s
plt.show()
```

# generatind data when gives a standard normal distribuition sigma and mean

```python
Xsn=np.random.randn(100)
sigma=5
mean=60
X=Xsn*sigma+mean# it mean max values whould lie between 55 and 65
```

```python
X=np.round(X)
```

```python
X
```

# Histogram

```python
plt.hist?
```

```python
plt.style.use('seaborn-whitegrid')
plt.hist(X,bins=10,alpha=0.8)# alpha for opacity 
plt.title(" Number of students vs marks with mean=60 and std.=5")
plt.xlabel( " marks ")
plt.ylabel(" freq of students")
plt.show()
```
